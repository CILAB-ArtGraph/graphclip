```python
styles = pd.read_csv("data/processed/styles.csv")
styles
```

```python
def ask_mistral(summary, style):
    return ollama.chat(
        model="mistral",
        messages=[
            {
                "role": "user",
                "content": f'Describe in one short sentence of 20 words a protorype artwork in the style of {style}. Do not refer explicitly to existing paintings, but focus on listing meaningful visual inputs. Do not include the number of words. Get the info from the following summary: \n"{summary}"',
            },
        ],
    )["message"]["content"]
```

```python
styles["summary"] = styles.apply(lambda x: ask_mistral(x.summary, x["style"]), axis=1)
```

```python
styles.to_csv("./data/processed/styles_mistral_prototype.csv", index=False)
```
