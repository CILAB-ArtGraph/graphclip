```python
styles = pd.read_csv("data/external/genres.csv").drop("id", axis=1)
styles
```

```python
def ask_mistral(summary):
    return ollama.chat(
        model="mistral",
        messages=[
            {
                "role": "user",
                "content": f'Sum up the following content regarding an artistic style in 50 words.\n Focus on visual aspects present in the description : \n"{summary}"',
            },
        ],
    )["message"]["content"]
```

```python
styles["summary"] = styles.apply(lambda x: ask_mistral(x.summary), axis=1)
```

```python
styles
```

```python
styles.to_csv("data/processed/genres_mistral.csv", index=False)
```
